```python
#1
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def delete_node(node):
    if node is None or node.next is None:
        return

    next_node = node.next
    node.val = next_node.val
    node.next = next_node.next

# Helper function to create a linked list from a list of values
def create_linked_list(values):
    dummy = ListNode()
    current = dummy
    for val in values:
        current.next = ListNode(val)
        current = current.next
    return dummy.next

# Helper function to convert a linked list to a list of values
def linked_list_to_list(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values

# Test cases
head1 = create_linked_list([4, 5, 1, 9])
node1 = head1.next  # Node with value 5
delete_node(node1)
output1 = linked_list_to_list(head1)
print(output1)  # Output: [4, 1, 9]

head2 = create_linked_list([4, 5, 1, 9])
node2 = head2.next.next  # Node with value 1
delete_node(node2)
output2 = linked_list_to_list(head2)
print(output2)  # Output: [4, 5, 9]
```

```python
#2
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def remove_elements(head, val):
    dummy = ListNode()
    dummy.next = head
    current = dummy

    while current.next:
        if current.next.val == val:
            current.next = current.next.next
        else:
            current = current.next

    return dummy.next

# Helper function to create a linked list from a list of values
def create_linked_list(values):
    dummy = ListNode()
    current = dummy
    for val in values:
        current.next = ListNode(val)
        current = current.next
    return dummy.next

# Helper function to convert a linked list to a list of values
def linked_list_to_list(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values

# Test case
head1 = create_linked_list([1, 2, 6, 3, 4, 5, 6])
val1 = 6
output1 = remove_elements(head1, val1)
print(linked_list_to_list(output1))  # Output: [1, 2, 3, 4, 5]

head2 = create_linked_list([])
val2 = 1
output2 = remove_elements(head2, val2)
print(linked_list_to_list(output2))  # Output: []

head3 = create_linked_list([7, 7, 7, 7])
val3 = 7
output3 = remove_elements(head3, val3)
print(linked_list_to_list(output3))  # Output: []
```

```python
#3
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def merge_sorted_lists(list1, list2):
    dummy = ListNode()
    current = dummy

    while list1 and list2:
        if list1.val <= list2.val:
            current.next = list1
            list1 = list1.next
        else:
            current.next = list2
            list2 = list2.next
        current = current.next

    if list1:
        current.next = list1
    if list2:
        current.next = list2

    return dummy.next

# Helper function to create a linked list from a list of values
def create_linked_list(values):
    dummy = ListNode()
    current = dummy
    for val in values:
        current.next = ListNode(val)
        current = current.next
    return dummy.next

# Helper function to convert a linked list to a list of values
def linked_list_to_list(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values

# Test case
list1 = create_linked_list([1, 2, 4])
list2 = create_linked_list([1, 3, 4])
output = merge_sorted_lists(list1, list2)
print(linked_list_to_list(output))  # Output: [1, 1, 2, 3, 4, 4]
```

```python
#4
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def detect_cycle(head):
    # Step 1: Check if there is a cycle using Floyd's algorithm
    slow = head
    fast = head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            break
    else:
        # No cycle, return None
        return None

    # Step 2: Find the starting node of the cycle
    slow = head

    while slow != fast:
        slow = slow.next
        fast = fast.next

    return slow

# Helper function to create a linked list from a list of values
def create_linked_list(values, pos):
    dummy = ListNode()
    current = dummy
    cycle_start = None

    for i, val in enumerate(values):
        current.next = ListNode(val)
        current = current.next

        if i == pos:
            cycle_start = current

    # Connect the tail to the cycle_start node to create a cycle
    if cycle_start:
        current.next = cycle_start

    return dummy.next

# Helper function to convert a linked list to a list of values
def linked_list_to_list(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values

# Test cases
head1 = create_linked_list([3, 2, 0, -4], 1)
output1 = detect_cycle(head1)
if output1:
    print(f"tail connects to node index {linked_list_to_list(output1).index(output1.val)}")
else:
    print("no cycle")

head2 = create_linked_list([1, 2], 0)
output2 = detect_cycle(head2)
if output2:
    print(f"tail connects to node index {linked_list_to_list(output2).index(output2.val)}")
else:
    print("no cycle")

head3 = create_linked_list([1], -1)
output3 = detect_cycle(head3)
if output3:
    print(f"tail connects to node index {linked_list_to_list(output3).index(output3.val)}")
else:
    print("no cycle")
```

```python
#5
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def remove_nth_from_end(head, n):
    dummy = ListNode(0)
    dummy.next = head
    fast = slow = dummy

    # Move the fast pointer n+1 steps ahead of the slow pointer
    for _ in range(n + 1):
        fast = fast.next

    # Move both pointers until the fast pointer reaches the end
    while fast:
        fast = fast.next
        slow = slow.next

    # Remove the nth node from the end
    slow.next = slow.next.next

    return dummy.next

# Helper function to create a linked list from a list of values
def create_linked_list(values):
    dummy = ListNode()
    current = dummy

    for val in values:
        current.next = ListNode(val)
        current = current.next

    return dummy.next

# Helper function to convert a linked list to a list of values
def linked_list_to_list(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values

# Test cases
head1 = create_linked_list([1, 2, 3, 4, 5])
output1 = remove_nth_from_end(head1, 2)
print(linked_list_to_list(output1))  # Output: [1, 2, 3, 5]

head2 = create_linked_list([1])
output2 = remove_nth_from_end(head2, 1)
print(linked_list_to_list(output2))  # Output: []

head3 = create_linked_list([1, 2])
output3 = remove_nth_from_end(head3, 1)
print(linked_list_to_list(output3))  # Output: [1]
```

```python
#6
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def left_shift_linked_list(head, k):
    if not head or k == 0:
        return head

    # Calculate the length of the linked list
    length = 1
    current = head
    while current.next:
        current = current.next
        length += 1

    # Handle cases where k is greater than the length of the linked list
    k = k % length

    if k == 0:
        return head

    # Find the new head node after left-shifting by k nodes
    new_head_pos = length - k
    current = head
    for _ in range(new_head_pos - 1):
        current = current.next

    new_head = current.next
    current.next = None

    # Update the pointers to create the new linked list
    current = new_head
    while current.next:
        current = current.next

    current.next = head

    return new_head

# Helper function to create a linked list from a list of values
def create_linked_list(values):
    dummy = ListNode()
    current = dummy

    for val in values:
        current.next = ListNode(val)
        current = current.next

    return dummy.next

# Helper function to convert a linked list to a list of values
def linked_list_to_list(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values

# Test case
head = create_linked_list([2, 4, 7, 8, 9])
output = left_shift_linked_list(head, 3)
print(linked_list_to_list(output))  # Output: [8, 9, 2, 4, 7]
```

```python
#7
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def remove_zero_sum_sublists(head):
    dummy = ListNode(0)
    dummy.next = head

    prefix_sum = 0
    prefix_sum_map = {0: dummy}

    current = dummy.next

    while current:
        prefix_sum += current.val

        if prefix_sum in prefix_sum_map:
            start = prefix_sum_map[prefix_sum]
            start.next = current.next
        else:
            prefix_sum_map[prefix_sum] = current

        current = current.next

    return dummy.next

# Helper function to create a linked list from a list of values
def create_linked_list(values):
    dummy = ListNode()
    current = dummy

    for val in values:
        current.next = ListNode(val)
        current = current.next

    return dummy.next

# Helper function to convert a linked list to a list of values
def linked_list_to_list(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values

# Test case
head = create_linked_list([1, 2, -3, 3, 1])
output = remove_zero_sum_sublists(head)
print(linked_list_to_list(output))  # Output: [3, 1]
```

```python
#8
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def odd_even_linked_list(head):
    if not head or not head.next:
        return head

    odd_head = head
    even_head = head.next
    even_start = even_head

    while even_head and even_head.next:
        odd_head.next = even_head.next
        odd_head = odd_head.next
        even_head.next = odd_head.next
        even_head = even_head.next

    odd_head.next = even_start

    return head

# Helper function to create a linked list from a list of values
def create_linked_list(values):
    dummy = ListNode()
    current = dummy

    for val in values:
        current.next = ListNode(val)
        current = current.next

    return dummy.next

# Helper function to convert a linked list to a list of values
def linked_list_to_list(head):
    values = []
    current = head
    while current:
        values.append(current.val)
        current = current.next
    return values

# Test case
head = create_linked_list([1, 2, 3, 4, 5])
output = odd_even_linked_list(head)
print(linked_list_to_list(output))  # Output: [1, 3, 5, 2, 4]
```
